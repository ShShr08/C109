```python
import statistics
import random
import pandas

r = []
for i in range(1,1000):
  d1 = random.randint(1,6)
  d2 = random.randint(1,6)
  r.append(d1+d2)
mean = sum(r)/len(r)
print(mean)
```

```python
df = pandas.read_csv('data.csv')
hl = df['Height(Inches)'].to_list()
wl = df['Weight(Pounds)'].to_list()
hm = statistics.mean(hl)
wm = statistics.mean(wl)
hme = statistics.median(hl)
wme = statistics.median(wl)
print("hmean = {}\n wmean = {}\n hmedian = {}\n wmedian = {}".format(hm, wm, hme, wme))
hsd = statistics.stdev(hl)
wsd = statistics.stdev(wl)
```

```python
median = statistics.median(r)
mode = statistics.mode(r)
sd = statistics.stdev(r)
print('{} {} {}'.format(median, mode, sd))
```

```python
a1, a2 = mean-sd, mean+sd 
b1, b2 = mean-(2*sd), mean+(2*sd)
c1, c2 = mean-(3*sd), mean+(3*sd)
lod1 = [result for result in r if result>a1 and result<a2]
lod2 = [result for result in r if result>b1 and result<b2]
lod3 = [result for result in r if result>c1 and result<c2]
print("Percentage of data in 1st deviation: {}\n Percentage of data in 2nd deviation: {}\n Percentage of data in 3st deviation: {}"
.format(len(lod1)*100/len(r), len(lod2)*100/len(r), len(lod3)*100/len(r)))
```

```python
x1, x2 = hm-hsd, hm+hsd
y1, y2 = hm-(2*hsd), hm+(2*hsd)
z1, z2 = hm-(3*hsd), hm+(3*hsd)
lod1 = [result for result in hl if result>x1 and result<x2]
lod2 = [result for result in hl if result>y1 and result<y2]
lod3 = [result for result in hl if result>z1 and result<z2]
print("Percentage of data in 1st deviation: {}\n Percentage of data in 2nd deviation: {}\n Percentage of data in 3st deviation: {}"
.format(len(lod1)*100/len(hl), len(lod2)*100/len(hl), len(lod3)*100/len(hl)))
```

```python
d1, d2 = wm-wsd, wm+wsd
e1, e2 = wm-(2*wsd), wm+(2*wsd)
f1, f2 = wm-(3*wsd), wm+(3*wsd)
lod1 = [result for result in wl if result>d1 and result<d2]
lod2 = [result for result in wl if result>e1 and result<e2]
lod3 = [result for result in wl if result>f1 and result<f2]
print("Percentage of data in 1st deviation: {}\n Percentage of data in 2nd deviation: {}\n Percentage of data in 3st deviation: {}"
.format(len(lod1)*100/len(wl), len(lod2)*100/len(wl), len(lod3)*100/len(wl)))
```
